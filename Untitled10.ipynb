```python

def generate_weather_data(n_days=1095):  # 3 years of data
    """Generate realistic weather data with seasonal patterns and missing values"""

    np.random.seed(42)  # For reproducibility

    # Create date range
    start_date = datetime(2021, 1, 1)
    dates = [start_date + timedelta(days=i) for i in range(n_days)]

    # Generate seasonal temperature patterns
    day_of_year = np.array([d.timetuple().tm_yday for d in dates])
    seasonal_temp = 20 + 15 * np.sin(2 * np.pi * (day_of_year - 80) / 365)

    # Add random variations and trends
    temp_variation = np.random.normal(0, 5, n_days)
    yearly_trend = np.linspace(0, 2, n_days)  # Slight warming trend

    # Calculate temperature
    temperature = seasonal_temp + temp_variation + yearly_trend

    # Generate correlated humidity (inverse relationship with temperature)
    humidity = 70 - 0.8 * (temperature - 20) + np.random.normal(0, 10, n_days)
    humidity = np.clip(humidity, 10, 95)

    # Generate wind speed
    wind_speed = np.abs(np.random.normal(15, 8, n_days))

    # Generate pressure (slightly correlated with temperature)
    pressure = 1013 + 0.5 * (temperature - 20) + np.random.normal(0, 15, n_days)

    # Generate precipitation (more likely in certain seasons)
    precip_prob = 0.3 + 0.2 * np.sin(2 * np.pi * (day_of_year - 80) / 365)
    precipitation = np.where(np.random.random(n_days) < precip_prob,
                           np.abs(np.random.exponential(5, n_days)), 0)

    # Generate cloud cover
    cloud_cover = np.random.beta(2, 3, n_days) * 100

    # Create DataFrame
    weather_data = pd.DataFrame({
        'date': dates,
        'temperature': temperature,
        'humidity': humidity,
        'wind_speed': wind_speed,
        'pressure': pressure,
        'precipitation': precipitation,
        'cloud_cover': cloud_cover
    })

    # Add some extreme weather events
    extreme_indices = np.random.choice(n_days, size=20, replace=False)
    weather_data.loc[extreme_indices[:10], 'temperature'] += np.random.uniform(10, 20, 10)  # Heat waves
    weather_data.loc[extreme_indices[10:], 'temperature'] -= np.random.uniform(10, 15, 10)  # Cold snaps

    # Introduce missing values strategically
    missing_indices = np.random.choice(n_days, size=int(0.05 * n_days), replace=False)
    weather_data.loc[missing_indices[:len(missing_indices)//3], 'humidity'] = np.nan
    weather_data.loc[missing_indices[len(missing_indices)//3:2*len(missing_indices)//3], 'wind_speed'] = np.nan
    weather_data.loc[missing_indices[2*len(missing_indices)//3:], 'pressure'] = np.nan

    # Add some outliers
    outlier_indices = np.random.choice(n_days, size=15, replace=False)
    weather_data.loc[outlier_indices[:8], 'wind_speed'] *= 3  # Extreme wind events
    weather_data.loc[outlier_indices[8:], 'precipitation'] *= 5  # Heavy rainfall events

    return weather_data

# Generate the dataset
df = generate_weather_data()
print(f"üìà Generated weather dataset with {len(df)} records")
print(f"üìÖ Date range: {df['date'].min()} to {df['date'].max()}")
```

```python

print("\n" + "="*60)
print("üîç PATTERNS, TRENDS, AND ANOMALIES DETECTION")
print("="*60)

# Time series decomposition for trend analysis
from scipy.signal import detrend

# Calculate yearly trend
df_yearly = df.groupby('year')['temperature'].mean()
if len(df_yearly) > 1:
    yearly_trend = np.polyfit(range(len(df_yearly)), df_yearly.values, 1)[0]
    print(f"üìà Yearly temperature trend: {yearly_trend:.3f}¬∞C per year")

# Seasonal pattern analysis
seasonal_pattern = df.groupby('day_of_year')['temperature'].mean()
plt.figure(figsize=(15, 6))
plt.plot(seasonal_pattern.index, seasonal_pattern.values, linewidth=2)
plt.title('Seasonal Temperature Pattern (Average by Day of Year)')
plt.xlabel('Day of Year')
plt.ylabel('Temperature (¬∞C)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Anomaly detection using statistical methods
def detect_anomalies_iqr(series, multiplier=1.5):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    return (series < lower_bound) | (series > upper_bound)

def detect_anomalies_zscore(series, threshold=3):
    z_scores = np.abs(stats.zscore(series.dropna()))
    return z_scores > threshold

# Detect temperature anomalies
temp_anomalies_iqr = detect_anomalies_iqr(df['temperature'])
temp_anomalies_zscore = detect_anomalies_zscore(df['temperature'])

print(f"üö® Temperature anomalies (IQR method): {temp_anomalies_iqr.sum()}")
print(f"üö® Temperature anomalies (Z-score method): {temp_anomalies_zscore.sum()}")

# Pattern analysis for extreme weather events
print("\nüå™Ô∏è Extreme Weather Event Patterns:")

# Heat wave detection (consecutive hot days)
df['heat_wave'] = (df['is_hot_day'].rolling(window=3).sum() >= 3).astype(int)
heat_waves = df['heat_wave'].sum()
print(f"Heat wave periods (3+ consecutive hot days): {heat_waves}")

# Cold snap detection (consecutive cold days)
df['cold_snap'] = (df['is_cold_day'].rolling(window=3).sum() >= 3).astype(int)
cold_snaps = df['cold_snap'].sum()
print(f"Cold snap periods (3+ consecutive cold days): {cold_snaps}")

# Analyze correlation between weather variables and extreme events
print("\nüìä Weather Variable Correlations with Extreme Events:")
extreme_correlations = pd.DataFrame({
    'Variable': ['humidity', 'wind_speed', 'pressure', 'cloud_cover'],
    'Hot_Day_Correlation': [df[var].corr(df['is_hot_day']) for var in ['humidity', 'wind_speed', 'pressure', 'cloud_cover']],
    'Cold_Day_Correlation': [df[var].corr(df['is_cold_day']) for var in ['humidity', 'wind_speed', 'pressure', 'cloud_cover']]
})

print(extreme_correlations.round(3))
```

```python

print("\n" + "="*60)
print("üîß OUTLIERS HANDLING AND DATA TRANSFORMATIONS")
print("="*60)

# Identify outliers using multiple methods
outlier_results = {}

for column in ['temperature', 'humidity', 'wind_speed', 'pressure']:
    # IQR method
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    iqr_outliers = ((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))).sum()

    # Z-score method
    z_scores = np.abs(stats.zscore(df[column].dropna()))
    zscore_outliers = (z_scores > 3).sum()

    outlier_results[column] = {
        'IQR_outliers': iqr_outliers,
        'Zscore_outliers': zscore_outliers
    }

outlier_df = pd.DataFrame(outlier_results).T
print("üéØ Outlier Detection Summary:")
print(outlier_df)

# Outlier treatment strategies
df_treated = df.copy()

# Strategy 1: Capping extreme values (for wind speed)
wind_95th = df['wind_speed'].quantile(0.95)
wind_5th = df['wind_speed'].quantile(0.05)
df_treated['wind_speed_capped'] = df['wind_speed'].clip(lower=wind_5th, upper=wind_95th)

# Strategy 2: Log transformation for skewed data (precipitation)
df_treated['precipitation_log'] = np.log1p(df['precipitation'])  # log1p handles zeros

# Strategy 3: Box-Cox transformation for normalization
from scipy.stats import boxcox

# Only apply to positive values
positive_temp = df['temperature'] - df['temperature'].min() + 1
df_treated['temperature_boxcox'], lambda_temp = boxcox(positive_temp)

# Standardization for modeling
scaler = StandardScaler()
features_to_scale = ['temperature', 'humidity', 'wind_speed', 'pressure']
df_treated[['temp_scaled', 'humid_scaled', 'wind_scaled', 'press_scaled']] = scaler.fit_transform(df[features_to_scale])

print("‚úÖ Data transformations applied:")
print("- Wind speed capping (5th-95th percentile)")
print("- Log transformation for precipitation")
print("- Box-Cox transformation for temperature")
print("- Standardization for key features")

# Compare distributions before and after treatment
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Original vs capped wind speed
axes[0,0].hist(df['wind_speed'], bins=30, alpha=0.7, label='Original', density=True)
axes[0,0].hist(df_treated['wind_speed_capped'], bins=30, alpha=0.7, label='Capped', density=True)
axes[0,0].set_title('Wind Speed: Original vs Capped')
axes[0,0].legend()

# Original vs log-transformed precipitation
axes[0,1].hist(df['precipitation'], bins=30, alpha=0.7, label='Original', density=True)
axes[0,1].hist(df_treated['precipitation_log'], bins=30, alpha=0.7, label='Log-transformed', density=True)
axes[0,1].set_title('Precipitation: Original vs Log-transformed')
axes[0,1].legend()

# Original vs Box-Cox temperature
axes[1,0].hist(df['temperature'], bins=30, alpha=0.7, label='Original', density=True)
axes[1,0].hist(df_treated['temperature_boxcox'], bins=30, alpha=0.7, label='Box-Cox', density=True)
axes[1,0].set_title('Temperature: Original vs Box-Cox')
axes[1,0].legend()

# Original vs standardized temperature
axes[1,1].hist(df['temperature'], bins=30, alpha=0.7, label='Original', density=True)
axes[1,1].hist(df_treated['temp_scaled'], bins=30, alpha=0.7, label='Standardized', density=True)
axes[1,1].set_title('Temperature: Original vs Standardized')
axes[1,1].legend()

plt.tight_layout()
plt.show()
```
